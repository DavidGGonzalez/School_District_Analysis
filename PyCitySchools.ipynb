```python
# Let's start cleaning student's names here
```

```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
# Iterate through prefixes and suffixes in prefixes_suffixes and replace them with "" in the the student_data_df["student_name"] column
for prefsuff in prefixes_suffixes:
    students_data_df["student_name"] = students_data_df["student_name"].str.replace(prefsuff,"")
```

```python
students_data_df.tail(20)
```

```python
# Combine both data frames by merging them using the school_name field their id
school_data_complete_df = pd.merge(students_data_df, schools_data_df, on=["school_name","school_name"])
school_data_complete_df.head()
```
